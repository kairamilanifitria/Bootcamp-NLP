# Get data from a single page

## Base code

```python
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup

# url = 'https://www.detik.com/tag/pendidikan'
base_url = 'https://www.detik.com/search/searchall?query='
query = 'pendidikan'

response = requests.get(base_url + query)

if response.status_code == 200:
  soup = BeautifulSoup(response.text, 'html.parser')

  berita_elements = soup.select('.list-content .list-content__item')

  for berita in berita_elements:
    judul = berita.select_one('.media__title').text.strip()
    url_berita = berita.select_one('a')['href']
    summary_str = berita.select_one('.media__desc')
    summary = summary_str.text.strip() if summary_str != None else np.nan

    print('Judul:',judul)
    print('URL Berita:', url_berita)
    print('Summary Berita:', summary)
    print('-'*50)

else:
  print('Gagal mengambil data. Status kode:', response.status_code)
```

## Save the data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_single_page(base_url, query):

  response = requests.get(base_url + query)

  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    berita_elements = soup.select('.list-content .list-content__item')

    list_judul = []
    list_url = []
    list_summary = []
    for berita in berita_elements:
      judul = berita.select_one('.media__title').text.strip()
      url_berita = berita.select_one('a')['href']
      summary_str = berita.select_one('.media__desc')
      summary = summary_str.text.strip() if summary_str != None else np.nan

      list_judul.append(judul)
      list_url.append(url_berita)
      list_summary.append(summary)

  else:
    pass

  data = pd.DataFrame({
      'judul':list_judul,
      'url':list_url,
      'summary':list_summary
  })

  return data

```

```python
df_single = get_single_page(base_url='https://www.detik.com/search/searchall?query=', query='chatgpt')
display(df_single.info())
display(df_single.head())
```

```python
import requests
from bs4 import BeautifulSoup
import numpy as np

def get_multiple_pages(base_url, query, sort_by, page_count):

    list_judul = []
    list_url = []
    list_kategori = []
    list_date = []
    list_summary = []

    for page in range(1, page_count+1):
        url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

        # Mengirim permintaan GET ke URL
        response = requests.get(url)
        # Mengecek status kode respons
        if response.status_code == 200:
            # Mengubah respons menjadi objek BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
            berita_elements = soup.select('.list-content .list-content__item')

            # Mengiterasi setiap elemen berita
            for berita in berita_elements:
                # Mendapatkan judul dan URL berita
                judul = berita.select_one('.media__title').text.strip()
                url_berita = berita.select_one('a')['href']

                summary_str = berita.select_one('.media__desc')
                summary = summary_str.text.strip() if summary_str != None else np.nan

                category_str = berita.select_one('.media__subtitle')
                category = category_str.text.strip() if category_str != None else np.nan

                date_str = berita.select_one('.media__date')
                date = date_str.text.strip() if date_str != None else np.nan

                list_judul.append(judul)
                list_url.append(url_berita)
                list_kategori.append(category)
                list_date.append(date)
                list_summary.append(summary)

        else:
            pass

    data = pd.DataFrame({
        'judul':list_judul,
        'url':list_url,
        'kategori':list_kategori,
        'tanggal':list_date,
        'summary':list_summary,

    })

    return data
```

# Get detail content

## Base code

```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://www.detik.com/search/searchall?'
query = 'pendidikan'
sort_by = 'time'
page_count = 1
siteid = 63

for page in range(1, page_count+1):
    url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page) + '&siteid=' + str(siteid)

    # Mengirim permintaan GET ke URL
    response = requests.get(url)
    print('Page:', page)
    # Mengecek status kode respons
    if response.status_code == 200:
        # Mengubah respons menjadi objek BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
        berita_elements = soup.select('.list-content .list-content__item')

        # Mengiterasi setiap elemen berita
        for berita in berita_elements:
            # Mendapatkan judul dan URL berita
            judul = berita.select_one('.media__title').text.strip()
            url_berita = berita.select_one('a')['href']

            summary_str = berita.select_one('.media__desc')
            summary_berita = summary_str.text.strip() if summary_str != None else np.nan

            category_str = berita.select_one('.media__subtitle')
            category = category_str.text.strip() if category_str != None else np.nan

            date_str = berita.select_one('.media__date')
            date = date_str.text.strip() if date_str != None else np.nan

            print('Judul:', judul)
            print('URL Berita:', url_berita)
            print('Kategori:', category)
            print('Waktu Berita:', date)
            print('Summary Berita:', summary_berita)

            response_detail = requests.get(url_berita)
            # Mengecek status kode respons halaman detail
            if response_detail.status_code == 200:
                soup_detail = BeautifulSoup(response_detail.text, 'html.parser')
                # Mencari elemen detail

                detail_berita_str = soup_detail.select_one('.itp_bodycontent')
                if detail_berita_str != None:
                  detail_berita = ' '.join([i.text.strip() for i in detail_berita_str.find_all('p')])
                else:
                  detail_berita = np.nan
                author = soup_detail.find('meta', attrs={'name':'author'})['content']

                print('Detail:', detail_berita)
                print('Penulis:', author)


            else:
                print('Gagal mengambil halaman detail. Status kode:', response.status_code)

            print('---')
    else:
        print('Gagal mengambil data. Status kode:', response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

def get_multiple_pages_detail(base_url, query, sort_by, page_count, siteid, get_detail=True):

    list_judul = []
    list_url = []
    list_kategori = []
    list_date = []
    list_summary = []
    list_detail_berita = []
    list_author = []

    for page in range(1, page_count+1):
        url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page) + '&siteid=' + str(siteid)

        # Mengirim permintaan GET ke URL
        response = requests.get(url)
        # Mengecek status kode respons
        if response.status_code == 200:
            # Mengubah respons menjadi objek BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
            berita_elements = soup.select('.list-content .list-content__item')

            # Mengiterasi setiap elemen berita
            for berita in berita_elements:
                # Mendapatkan judul dan URL berita
                judul = berita.select_one('.media__title').text.strip()
                url_berita = berita.select_one('a')['href']

                summary_str = berita.select_one('.media__desc')
                summary_berita = summary_str.text.strip() if summary_str != None else np.nan

                category_str = berita.select_one('.media__subtitle')
                category = category_str.text.strip() if category_str != None else np.nan

                date_str = berita.select_one('.media__date')
                date = date_str.text.strip() if date_str != None else np.nan


                list_judul.append(judul)
                list_url.append(url_berita)
                list_kategori.append(category)
                list_date.append(date)
                list_summary.append(summary)
                # print(url_berita)
                if get_detail:
                    response_detail = requests.get(url_berita)
                    # Mengecek status kode respons halaman detail
                    if response_detail.status_code == 200:
                        soup_detail = BeautifulSoup(response_detail.text, 'html.parser')

                        # Mencari elemen detail
                        detail_berita_str = soup_detail.select_one('.itp_bodycontent')
                        if detail_berita_str != None:
                          detail_berita = ' '.join([i.text.strip() for i in detail_berita_str.find_all('p')])
                        else:
                          detail_berita = np.nan
                        author = soup_detail.find('meta', attrs={'name':'author'})['content']

                        list_detail_berita.append(detail_berita)
                        list_author.append(author)

        else:
            pass

    data = pd.DataFrame({
        'judul':list_judul,
        'url':list_url,
        'kategori':list_kategori,
        'tanggal':list_date,
        'summary':list_summary,
        'isi_berita':list_detail_berita,
        'penulis':list_author,

    })

    return data
```

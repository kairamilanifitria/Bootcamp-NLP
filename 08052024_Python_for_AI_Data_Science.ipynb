## pandas manipulation

```python
# 1.mengatasi missing values dengan drop data

df
```

```python
df.dropna()  # membuang data yang kosong/hilang
```

```python
# 2. mengisi dengan nilai rata-rata untuk field yang kosong

df
```

```python
x = df['Python Score'].mean()
```

```python
df['Python Score'].fillna(x, inplace=True)
```

```python
df

# nanti akan terisi oleh nilai dari hasil rata-rata pada field yang kosong
```

```python
# 3. mengisi data kosong dengan mode()

df
```

```python
x = df['Average Score'].mode()
x
```

```python
df['Average Score'] = df['Average Score'].fillna(x)
```

```python
df['Average Score'].fillna(x, inplace=True)
```

```python
df
```

```python
df.loc[3, 'Average Score'] = 80
```

```python
df
```

```python
# mengubah nilai dengan pengkondisian if

for x in df.index:
  if df.loc[x, 'Python Score'] > 100:
    df.loc[x, 'Python Score'] = 100.0
```

```python
df
```

### pandas plotting

```python
import matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

# Plot skor python
plt.subplot(1,2,1)
plt.bar(df['Name'], df['Python Score'], color='skyblue')
plt.title('Python Score')
plt.xlabel('Name')
plt.ylabel('Score')

# Plot skor rata-rata
plt.subplot(1,2,2)
plt.bar(df['Name'], df['Average Score'], color='lightgreen')
plt.title('Average Score')
plt.xlabel('Name')
plt.ylabel('Score')

plt.tight_layout()
```
